# Tungsten qualification simulator walkthrough

This notebook demonstrates the stylised tungsten irradiation simulator used in the `material-qualification-demo` repository.

```
from material_qualification_demo.simulators.tungsten import (
    TungstenQualificationConfig,
    TungstenQualificationSimulator,
)
import pandas as pd
```

```
config = TungstenQualificationConfig()
sim = TungstenQualificationSimulator(config)
anchor_points = sim.anchor_points(
    dpa_bounds=(0.0, 3.0),
    temperature_bounds=(300.0, 1400.0),
    impurity_bounds=(0.001, 0.02),
)
anchor_points
```

```
outputs = sim.forward(anchor_points)
pd.DataFrame(outputs)
```

## Notes

- Yield stress and hardness increase with irradiation damage.
- Thermal diffusivity decreases with damage and impurities.
- High irradiation temperature partially recovers damage.
